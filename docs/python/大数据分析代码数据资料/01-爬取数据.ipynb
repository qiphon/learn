```python
import urllib
from urllib import request
import time
import datetime
import json
import csv
import codecs
import re

nowTime = lambda:str(int(round(time.time() * 1000)))

class EarthquakeCrawler():
    """
    数据接口
    """
    __interface = 'http://www.ceic.ac.cn/ajax/search?page=%s&&start=&&end=&&jingdu1=&&jingdu2=&&weidu1=&&weidu2=&&height1=&&height2=&&zhenji1=&&zhenji2=&&callback=callback&_=%s'

    """
    数据文件名
    """
    __filename = 'earth_quake.csv'
    
    """
    抓取数据
    """
    def crawl(self):
        now = nowTime()
        data = []
        
        url = self.__interface % (1, now)
        # 处理第一页数据并获取总页数
        page = self.fetch_page(url) 
        pdObj = self.parse_data(page)
        data = pdObj['shuju']
        num = pdObj['num']
        print('总共 %s 页' % num)
        head = data[0].keys()
        
        # 准备写入文件文件
        f = codecs.open( self.__filename, 'a', 'utf_8_sig')#解决写入csv时中文乱码
        writer = csv.DictWriter(f,head)

        writer.writeheader()    # 表头
        writer.writerows(data)  # 数据
        
        # 循环抓取剩下的数据
        n = 2
        while n <= num:
            
            url = self.__interface % (n, now)
            time.sleep(0.1)
            # 处理数据
            page = self.fetch_page(url) 
            pdObj = self.parse_data(page)
            data = pdObj['shuju']

            writer.writerows(data)
            # 显示进度
            if n % 10 == 0 :
                print( '[%s]' % n)
            else :
                print('.', end='')
            n += 1
            
        f.close()


    """
    抓取单页数据
    """
    def fetch_page(self, url):
        req = request.Request(url)
        res = request.urlopen(req)
        return res.read().decode(encoding='utf-8')#.encode('utf-8').decode(encoding='utf-8')
    """
    解析数据
    """
    def parse_data(self, data):
        reg_data = r'callback\((.*)\)'
        data_str = re.search(reg_data, data).group(1)        
        dataObj = json.loads(data_str)
        return dataObj
        
        
if __name__ == '__main__':
    crawler = EarthquakeCrawler()
    crawler.crawl()
    print('抓取完毕！  ')
```
